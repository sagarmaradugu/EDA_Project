```python
m1=titanic.Age.mean()
```

```python
titanic['Age']=titanic.Age.fillna(m1)
```

```python
m2=titanic.Embarked.value_counts().idxmax()
```

```python
titanic['Embarked']=titanic.Embarked.fillna(m2)
```

```python
titanic=titanic.drop(['PassengerId','Name','Ticket', 'Fare', 'Cabin'],axis=1)
```

```python
x=titanic.drop('Survived',axis=1)
```

```python
x=pd.get_dummies(x).values
```

```python
y=titanic['Survived'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
predicted_value = model.predict(xtest)
```

```python
predicted_value
```

```python
from sklearn.metrics import confusion_matrix,recall_score,precision_score,f1_score,accuracy_score
```

```python
confusion_matrix(ytest,predicted_value)
```

```python
recall_score(ytest,predicted_value,pos_label = 0)
```

```python
precision_score(ytest,predicted_value,pos_label=0)
```

```python
f1_score(ytest,predicted_value,pos_label=0)
```

```python
accuracy_score(ytest,predicted_value)
```
